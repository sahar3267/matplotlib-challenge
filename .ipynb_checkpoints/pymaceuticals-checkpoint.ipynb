```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load
mouse_drug_data_to_load = os.path.join("Pymaceuticals", "data", "mouse_drug_data.csv")
clinical_trial_data_to_load = os.path.join("Pymaceuticals", "data", "clinicaltrial_data.csv")

# Read the Mouse and Drug Data and the Clinical Trial Data
df_mouse_drug_data = pd.read_csv(mouse_drug_data_to_load )
df_clinical_trial_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
data = pd.merge(df_mouse_drug_data, df_clinical_trial_data, on="Mouse ID")


# Display the data table for preview
data.head()
```

```python
data.columns
```

# Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
avg_tumor_volume = data.groupby(['Drug', 'Timepoint'])["Tumor Volume (mm3)"].mean().reset_index()
# Convert to DataFrame
avg_tumor_volume = pd.DataFrame(avg_tumor_volume)
# Preview DataFrame    
avg_tumor_volume
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
sem_tumor_volume = data.groupby(['Drug', 'Timepoint'])["Tumor Volume (mm3)"].sem().reset_index()

# Convert to DataFrame
sem_tumor_volume = pd.DataFrame(sem_tumor_volume)

# Preview DataFrame
sem_tumor_volume.head(10)
```

```python
# Minor Data Munging to Re-Format the Data Frames
avg_tumor_volume_pivot = avg_tumor_volume.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')
sem_tumor_volume_pivot = sem_tumor_volume.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')

# Preview that Reformatting worked
avg_tumor_volume_pivot.head()
# sem_tumor_volume_pivot.head()
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(avg_tumor_volume_pivot.index, avg_tumor_volume_pivot["Capomulin"], yerr=sem_tumor_volume_pivot["Capomulin"],marker="o",linestyle="dashed")
plt.errorbar(avg_tumor_volume_pivot.index, avg_tumor_volume_pivot["Placebo"], yerr=sem_tumor_volume_pivot["Placebo"])
plt.errorbar(avg_tumor_volume_pivot.index, avg_tumor_volume_pivot["Ketapril"], yerr=sem_tumor_volume_pivot["Ketapril"])
plt.errorbar(avg_tumor_volume_pivot.index, avg_tumor_volume_pivot["Infubinol"], yerr=sem_tumor_volume_pivot["Infubinol"])
plt.title("Tumor Response to Treatment")
plt.ylabel("Tumor Volume (mm3)")
plt.xlabel("Time(Day)")
plt.grid()
# Save the Figure
# plt.savefig("name_of_directory/name_of_the_file.png")
# Show the Figure
plt.show()
```

Observable Trend Concerning Average Tumor Volume: The mice that were given Capomulin had a decrease in the avereage tumor volume over the 45 days. Whereas, the mice that were given Infubinol, Ketapril, and the Placebo had an increase in the average tumor volume over the 45 days. This may suggest that Capomulin decreases tumor volume whereas the other two drugs, Infubinol, Ketapril had little to no effect on average tumor volume since they both behaved very similiar to the placebo. In fact, since the error bars overlap the Placebo error bars for both the drugs, Inubinol and Ketapril, there is not a statistically significantly different result from the two drugs and the placebo over the 45 days.

# Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint
avg_metastatic_sites = data.groupby(['Drug', 'Timepoint'])["Metastatic Sites"].mean().reset_index()
# Convert to DataFrame
avg_metastatic_sites = pd.DataFrame(avg_metastatic_sites)
# Preview DataFrame
avg_metastatic_sites.head(10)
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Tim
sem_metastatic_sites = data.groupby(['Drug', 'Timepoint'])["Metastatic Sites"].sem().reset_index()
# Convert to DataFrame
sem_metastatic_sites = pd.DataFrame(sem_metastatic_sites)
# Preview DataFrame
sem_metastatic_sites
```

```python
# Minor Data Munging to Re-Format the Data Frames
avg_metastatic_sites_pivot = avg_metastatic_sites.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')
sem_metastatic_sites_pivot = sem_metastatic_sites.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')


# Preview that Reformatting worked
avg_metastatic_sites_pivot.head()
#sem_metastatic_sites_pivot.head()
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(avg_metastatic_sites_pivot.index, avg_metastatic_sites_pivot["Capomulin"], yerr=sem_metastatic_sites_pivot["Capomulin"],marker="o",linestyle="dashed")
plt.errorbar(avg_metastatic_sites_pivot.index, avg_metastatic_sites_pivot["Ketapril"], yerr=sem_metastatic_sites_pivot["Ketapril"],marker="o",linestyle="dashed")
plt.errorbar(avg_metastatic_sites_pivot.index, avg_metastatic_sites_pivot["Infubinol"], yerr=sem_metastatic_sites_pivot["Infubinol"],marker="o",linestyle="dashed")
plt.errorbar(avg_metastatic_sites_pivot.index, avg_metastatic_sites_pivot["Placebo"], yerr=sem_metastatic_sites_pivot["Placebo"],marker="o",linestyle="dashed")
plt.title("Metastatic Response to Treatment")
plt.ylabel("Metastatic Sites")
plt.xlabel("Treatment duration(Days)")
plt.grid()
# Save the Figure
# plt.savefig("name_of_directory/name_of_the_file.png")
# Show the Figure
plt.show()
```

Observable Trend Concerning the Average of Metastatic Sites: While all treatments saw an increase in average number of metastatic sites over the 45 days, with the Placebo having the highest average metastatic sites at all measured times. But, by day 45 only Capomulin and Infubinol had statistically significant lower average metastatic sites than the placebo. This would suggest that both Capomulin and Infubinol may reduce the average number of metastatic sites than being unmedicated, while Ketapril had statistically no effect.

# Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
survival_count = data.groupby(['Drug', 'Timepoint'])["Tumor Volume (mm3)"].count()
# Convert to DataFrame
survival_count = pd.DataFrame({"mouse_count": survival_count}).reset_index()
# Preview DataFrame
survival_count.head(10)

```

```python
# Minor Data Munging to Re-Format the Data Frames
mous_count= survival_count.pivot(index="Timepoint", columns="Drug")["mouse_count"]
# Preview the Data Frame
mous_count.head()
```

```python
# Generate the Plot (Accounting for percentages)
plt.errorbar(mous_count.index, mous_count["Capomulin"], yerr=sem_metastatic_sites_pivot["Capomulin"],marker="o",linestyle="dashed")
plt.errorbar(mous_count.index, mous_count["Ketapril"], yerr=sem_metastatic_sites_pivot["Ketapril"],marker="o",linestyle="dashed")
plt.errorbar(mous_count.index, mous_count["Infubinol"], yerr=sem_metastatic_sites_pivot["Infubinol"],marker="o",linestyle="dashed")
plt.errorbar(mous_count.index, mous_count["Placebo"], yerr=sem_metastatic_sites_pivot["Placebo"],marker="o",linestyle="dashed")
plt.title("Survival during Treatment")
plt.ylabel("Survival Rates(%)")
plt.xlabel("Time(Days)")
plt.grid()
# Save the Figure

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
percentage_change = (avg_tumor_volume_pivot.iloc[-1]/(avg_tumor_volume_pivot.iloc[0])-1)*100
# Display the data to confirm
percentage_change

```

```python

# Store all Relevant Percent Changes into a Tuple

percentage_change = percentage_change.sort_values()

# Splice the data between passing and failing drugs
failingDrug_s = percentage_change[percentage_change < 0].sort_values()
passingDrug_s = percentage_change.copy()
passingDrug_s.loc[passingDrug_s.values < 0] = 0
passingDrug_s = passingDrug_s.sort_values()

# Orient widths. tick marks, etc. 
x_axis = np.arange(len(percentage_change))
tick_locations = [value+0.5 for value in x_axis]

plt.figure(figsize=(14,4))
# Use functions to label the percentages of changes
ax = failingDrug_s.plot(kind='bar', color='g', width=1)
passingDrug_s.plot(kind='bar', color='r', width=1, ax=ax)
plt.xticks(tick_locations, percentage_change.index)

# Add labels
plt.title("Tumor Change Over 45 Days Treatment")
plt.ylabel("% Tumor Volume Change")
plt.xlim(-0.7,len(percentage_change)-0.3)
plt.ylim(-30,70)
plt.grid()


# For each bar: Place a label % for negative & positive
rects = ax.patches
for rect in rects:
    # Get X and Y placement of label from rect.
    y_value = rect.get_height()
    if y_value < 0.0:
        x_value = rect.get_x() + rect.get_width() / 2
        percentY = str(int(y_value)) + '%'
        plt.text(x_value-0.1, -5, percentY, color='w')
    elif y_value > 0.0:
        x_value = rect.get_x() + rect.get_width() / 2
        percentY = str(int(y_value)) + '%'
        plt.text(x_value-0.1, 5, percentY, color='w')
        
# Save the Figure
#plt.savefig("plots/Tumor_Change_Over_45_Days_Treatment2.png")
plt.show()
```
